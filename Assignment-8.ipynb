```python
#
import requests
import csv

class ConvertUnit:

    def __init__(self,rate,inputValue):
        self.r = rate
        self.i = inputValue
    
    def convert(self):
        return self.i/self.r

response = requests.get('https://api.frankfurter.app/latest?amount=1&from=USD&to=EUR')

exchangeRate = response.json()
ratE = exchangeRate['rates']['EUR']



inputValue = float(input("Enter USD value:"))

converto = ConvertUnit(ratE,inputValue)
print(f'The Euro value is: {converto.convert()}')

```

```python
response1 = requests.get('https://www.metaweather.com/api/location/1521894/')
weather = response1.json()

with open('weatherData.csv','w',newline = '') as f: #newline = '' to stop making a new line while starting writing
    writer = csv.DictWriter(f, fieldnames=["Temp","Humidity",'Visibility','Air_pressure','Wind_speed','Date','Time'])
    writer.writeheader()
    for e in weather['consolidated_weather']:
        Xtemp = e["the_temp"]
        Xhum = e["humidity"]
        Xvisi = e['visibility']
        XairPress = e['air_pressure']
        Xwind = e['wind_speed']
        Xdate = e['applicable_date']
        Xtime = e['created']
        writer.writerow(
            {
                "Temp" : Xtemp,
                "Humidity" : Xhum,
                'Visibility' : Xvisi,
                'Air_pressure' : XairPress,
                'Wind_speed' : Xwind,
                'Date' : Xdate,
                'Time' : Xtime

            }
        )

import pandas as pd

data = pd.read_csv('weatherData.csv')
data
```

```python
response2 = requests.get('https://jsonplaceholder.typicode.com/users')
usersData = response2.json()

with open('usersData.csv','w',newline = '') as f: #newline = '' to stop making a new line while starting writing
    writer = csv.DictWriter(f, fieldnames=["name","username","email","street","suite","city","zipcode","lat","lng"])
    writer.writeheader()
    for e in usersData:
        Xname = e["name"]
        Xuser = e["username"]
        Xemail = e['email']
        Xstreet = e['address']['street']
        Xsuite = e['address']['suite']
        Xcity= e['address']['city']
        Xzip = e['address']['zipcode']
        Xlat = e['address']['geo']['lat']
        Xlng = e['address']['geo']['lng']
        writer.writerow(
            {
                "name" : Xname,
                "username" : Xuser,
                'email' : Xemail,
                'street' : Xstreet,
                'suite' : Xsuite,
                'city' : Xcity,
                'zipcode' : Xzip,
                'lat' : Xlat,
                'lng' : Xlng

            }
        )
import pandas as pd1

data1 = pd1.read_csv('usersData.csv')
data1
```
